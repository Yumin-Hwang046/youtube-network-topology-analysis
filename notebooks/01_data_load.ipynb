## 3) Download Data

```python
import os
import urllib.request
import gzip
import shutil

data_dir = "../data/raw"
os.makedirs(data_dir, exist_ok=True)
target_file = os.path.join(data_dir, "com-youtube.ungraph.txt")
gz_file = target_file + ".gz"

if not os.path.exists(target_file):
    print("Downloading YouTube graph from Stanford SNAP...")
    urllib.request.urlretrieve("https://snap.stanford.edu/data/bigdata/communities/com-youtube.ungraph.txt.gz", gz_file)
    print("Extracting dataset...")
    with gzip.open(gz_file, 'rb') as f_in:
        with open(target_file, 'wb') as f_out:
            shutil.copyfileobj(f_in, f_out)
    os.remove(gz_file)
    print("Dataset ready.")
else:
    print("Dataset already exists. Ready for analysis.")
```

# Data Loading

```python
import networkx as nx

# Load the graph from the txt file
G = nx.read_edgelist(
    "../data/raw/com-youtube.ungraph.txt",
    comments="#",
    nodetype=int
)

print("Nodes:", G.number_of_nodes())
print("Edges:", G.number_of_edges())
```
