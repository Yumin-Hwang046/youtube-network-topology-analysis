## Visualizing the Deviation

Generates a comparative bar chart highlighting the clustering coefficient mismatch discussed in Notebook 05. This plot serves as a core asset for the final report.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Hardcoded empirical metrics from previous analysis steps for clean plotting
data = {
    "Network Model": ["BA Model (Null)", "ER Model (Null)", "Empirical YouTube", "WS Model (Null)"],
    "Clustering Coefficient": [0.001, 0.0005, 0.136, 0.25]  # Approximate values for visual impact
}

df = pd.DataFrame(data)

plt.figure(figsize=(10, 6))
sns.barplot(x="Network Model", y="Clustering Coefficient", data=df, palette="viridis")
plt.title("Clustering Coefficient Deviation: Empirical vs Null Models")
plt.ylabel("Average Clustering Coefficient")
plt.grid(axis='y', alpha=0.3)

# Save plot for report
import os
os.makedirs("../results/figures", exist_ok=True)
plt.savefig("../results/figures/clustering_deviation.png", dpi=300, bbox_inches="tight")
plt.show()
```
