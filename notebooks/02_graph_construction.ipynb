## 1) Graph Definition

Before constructing the graph, we must clearly define what the network represents:
- **Nodes ($V$)**: Individual YouTube users.
- **Edges ($E$)**: Mutual friendships between users. The graph is **undirected** and **unweighted**.

```python
import networkx as nx

# Load the graph from the txt file
G = nx.read_edgelist(
    "../data/raw/com-youtube.ungraph.txt",
    comments="#",
    nodetype=int
)

print("Nodes:", G.number_of_nodes())
print("Edges:", G.number_of_edges())
```

---
## 2) Graph Summary and Metadata

Verify the fundamental structure of the constructed NetworkX object.

```python
def print_graph_summary(G):
    """
    Prints basic metadata and structural properties of the graph.
    """
    print("--- Graph Summary ---")
    print(f"Type: {type(G).__name__}")
    print(f"Directed: {G.is_directed()}")
    print(f"Number of Nodes: {G.number_of_nodes():,}")
    print(f"Number of Edges: {G.number_of_edges():,}")
    print("---------------------")

print_graph_summary(G)
```

---
## 3) Export Logic

To maintain modularity across notebooks, save this construction logic into `src/graph_utils.py`. Subsequent notebooks can load the graph cleanly without repeating code.

```python
%%writefile ../src/graph_utils.py
import networkx as nx
import os

def load_youtube_graph(filepath="../data/raw/com-youtube.ungraph.txt"):
    """
    Loads the YouTube graph from the standard data directory.
    """
    if not os.path.exists(filepath):
        raise FileNotFoundError(f"Dataset not found at {filepath}. Please run 01_data_load.ipynb first.")
        
    print(f"Loading graph from {filepath}...")
    G = nx.read_edgelist(
        filepath,
        comments="#",
        create_using=nx.Graph(),
        nodetype=int
    )
    return G
```

> **Interpretation (Network Science Perspective):**
> 1. **Result Meaning**: The graph is successfully parsed as an undirected network, reflecting mutual friendships without hierarchical directionality.
> 2. **Structural Importance**: Establishing the foundational $N$ (nodes) and $E$ (edges) is critical. The massive scale ($N > 10^6$) dictates that exact path-based metrics will be computationally intractable, necessitating sampling approaches in subsequent steps.
> 3. **Next Step**: With the topology defined, we must extract its macroscopic properties (degree distribution, clustering) to formulate a structural hypothesis.
